```python
# Import pandas
import pandas as pd
```

```python
# Load the file into pandas dataframe
df = pd.read_excel('board2.xlsx')
```

## 1.A

```python
# Filter the students with Instrumentation Track from Luzon
instru = df[(df['Track'] == 'Instrumentation') & (df['Electronics'] > 70)].copy()
instru['Hometown'] = 'Luzon'
```

```python
# Prints the data of the students with the given filters
print("\nInstru DataFrame:")
print(instru[['Name', 'Track', 'GEAS', 'Electronics', 'Hometown']])
```

## 1.B

```python
# Filters students from Mindanao
mindy = df[(df['Gender'] == 'Female')].copy()
```

```python
# Computes the average scores of the students
mindy.loc[:, 'Average'] = mindy[['Math', 'GEAS', 'Electronics', 'Communication']].mean(axis=1)
# Filters the students with average scores equal to or greater than 55
mindy = mindy[mindy['Average'] >= 55]
mindy['Hometown'] = 'Mindanao'
```

```python
# Prints the data of the students with the given filters
print("\nMindy DataFrame:")
print(mindy[['Name', 'Track', 'Electronics', 'Average', 'Hometown']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Average'] = df[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1)
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(x='Track', y='Average', hue='Gender', data=df)
plt.title('Features contribution to average grade')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(x = 'Hometown', y = 'Average', data=df)
plt.title('Hometown contribution to average score')
plt.xlabel('Hometown')
plt.ylabel('Average Score')
plt.show()
```
